```python
def create_kto_data(yyyy, mm): # 년월 받아서 엑셀파일 오픈
    # 1. 엑셀파일 이름, 경로 결정
    file_path = f'./raw/kto_{yyyy+mm}.xlsx'

    # 2. 엑셀파일 불러오기
    df = pd.read_excel(file_path
                      , header=1
                      , usecols='A:G'
                      , skipfooter=4)
    # 3. 기준년월 컬럼 추가
    df['기준년월'] = f'{yyyy}-{mm}'
    # 4. 국적 컬럼에 대륙제거 국가만 남기기
    contients_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']
    condition = df['국적'].isin(contients_list) == False
    df_country = df[condition].reset_index(drop=True)
    # 5. 대륙 컬럼 추가
    contients = ['아시아'] * 25 +['아메리카'] * 5 +['유럽'] * 23 +['오세아니아'] * 3 +['아프리카'] * 2 + \
                ['기타'] + ['교포']
    df_country['대륙'] = contients
    # 6. 국가별 관광객비율(%) 컬럼추가
    df_country['관광객비율(%)'] = round((df_country['관광'] / df_country['계'])*100,1)
    # 7. 전체비율(%) 추가
    tourist_sum = sum(df_country['관광'])
    df_country['전체비율(%)'] = \
    round((df_country['관광'] / tourist_sum) * 100, 1)
    # 8. 결과출력
    return df_country

create_kto_data('2010', '02')
```

##### 반복문으로 125개 엑셀 하나로 합치기

```python
# kto_yyyymm.xlsx
df = pd.DataFrame()
temps = []

for yyyy in range(2010, 2021): #2020년까지니깐 +1 -> 2021
    for mm in range(1, 13):
        mm_str = str(mm).zfill(2)
        yyyymm = f'{yyyy}{mm_str}'
        if(yyyymm == '202006'): break
        # df.append() 함수는 없어짐.
        temp = create_kto_data(str(yyyy), mm_str)
        temps.append(temp)

        # print(yyyymm, temp.size, sep='|')

print(len(temps))
df = pd.concat(temps)
df.reset_index()
```

```python
df = df.reset_index(drop=True)
df.tail()
```

```python
# 최종 결과 저장
df.to_excel('./result/kto_total.xlsx', index=False)
```
