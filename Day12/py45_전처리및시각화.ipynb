```python
# import
import pandas as pd
import numpy as np
```

```python
# 스타벅스 매장 리스트
sbuck_df = pd.read_excel('./seoul_starbucks.xlsx', header=0)
sbuck_df.head()
```

```python
sbuck_df['주소'][1].split()[1]
```

```python
gu_list = []
for item in sbuck_df['주소']:
    gu = item.split()[1]
    gu_list.append(gu)

sbuck_df['구'] = gu_list
sbuck_df
```

#### 서울시 행정구역별 위경도 전처리
 - 대한민국 행정구역 위경도 엑셀파일에서 서울만 필터링

```python
seoul_loc_df = pd.read_excel('./대한민국_행정구역_위경도.xlsx')
seoul_loc_df.head()
```

```python
seoul_gu_df = seoul_loc_df[seoul_loc_df['do'] == '서울']
seoul_gu_df.head()
```

```python
seoul_gu_df.info()
```

```python
seoul_gu_df.to_excel('./seoul_gu_info.xlsx', index=False)
```

```python
sbucks_gu_counts = sbuck_df.pivot_table(index='구',
                     values='매장명',
                     aggfunc='count').rename(columns= {'매장명':'스타벅스_매장수'})
sbucks_gu_counts.head()
```
